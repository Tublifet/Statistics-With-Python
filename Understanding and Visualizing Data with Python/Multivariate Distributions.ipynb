# Mutivariate Distributions in Python
Sometimes, we can get a lot of information about how two variables (or more) relate if we plot them together. This tutorial aims to show how plotting two variables together can get us information that plotting each one separately may miss

```python
# import the packages we are going to be using 
import numpy as np # for getting our distribution
import matplotlib.pyplot as plt # for plotting
import seaborn as sns; sns.set() # for a different plotting theme

# Just know if we have a rho of 1 then we will get a perfectly
# upward slopping line, and if we have a rho of -1, we will get
# a perfectly downward slopping line. A rho of 0 will
# get us a 'cloud' of points scattered around the mean

r = 1

mean = [15, 5]
cov = [[1, r], [r, 1]]
x, y = x, y = np.random.multivariate_normal(mean, cov, 400).T

# Adjust the figure size
plt.figure(figsize = (10,5))

# Plot the histograms of X and Y next to each other
plt.subplot(1,2,1)
plt.hist(x = x, bins = 15)
plt.title("X")

plt.subplot(1,2,2)
plt.hist(x = y, bins = 15)
plt.title("Y")

plt.show()
```

```python
# Plot the data
plt.figure(figsize = (10,10))
plt.subplot(2,2,2)
plt.scatter(x = x, y = y)
plt.title("Joint Distribution of X and Y")

# Plot the Marginal Distribution of X
plt.subplot(2,2,4)
plt.hist(x = x, bins = 15)
plt.title("Marginal Distribution of X")

# Plot the Marginal Distribution of Y
plt.subplot(2,2,1)
plt.hist(x = x, orientation = "horizontal" ,bins = 15)
plt.title("Marginal Distribution of Y")

plt.show()
```
