# Visualizing Data in Python

## Tables, Histogram, Boxplots and Slicing for Statistics

When working with a new dataset, one of the most useful thing to do is to begin to visualizing the data. By using tables, Histograms, box plots, and other visual tools, we can get a better idea of what the data maybe trying to tell us, and we can gain insights into the data that we may have not discovered otherwise.


```python
# We first need to import the packages that we will be using

import seaborn as sns # for plotting
import matplotlib.pyplot as plt # for showing plots

# load in the data set
tips_data = sns.load_dataset("tips")
```

# Visualizing the Data - Tables

```python
# Print out the first few rows of the data
tips_data.head()
```

# Describing Data

```python
# Print out the summary statistics for the quantitative variables
tips_data.describe()
```

# Creating a Histogram

```python
# Plot a histogram of the total bill
sns.displot(tips_data["total_bill"], kde = False).fig.suptitle("Histogram of Total Bill")
plt.show()
```

```python
# Plot a histogram of the tips
sns.displot(tips_data["tip"], kde = False).fig.suptitle("Histogram of Total Tip")
plt.show()
```

```python
# Plot a histogram of both
sns.displot(tips_data["tip"], kde = False)
sns.displot(tips_data["total_bill"], kde = False).fig.suptitle('Histogram of both Total Tip and Total Bill')
plt.show()
```

# Create a Boxplot

```python
# Create a Boxplot of the total bill
sns.boxplot(tips_data["total_bill"]).set_title("Box plot of the Total Bill")
plt.show()
```

```python
# Create a boxplot of the tip
sns.boxplot(tips_data["tip"]).set_title("Box plot of Total Tip")
plt.show()
```

```python
# Create a boxplot of both
sns.boxplot(tips_data["total_bill"])
sns.boxplot(tips_data["tip"]).set_title("Box plot of both")

plt.show()
```

# Creating Histogram and Boxplots Plotted by Groups

```python
# Plotting tips grouped by smoking
sns.boxplot(x = tips_data["tip"], y = tips_data["smoker"])

g = sns.FacetGrid(tips_data, row = "smoker")
g = g.map(plt.hist, "tip")

plt.show()
```

```python
# Plotting tips grouped by time of day
sns.boxplot(x = tips_data["tip"], y = tips_data["time"])

g = sns.FacetGrid(tips_data, row = "time")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
# Plotting tips grouped by day of week
sns.boxplot(x = tips_data["tip"], y = tips_data["day"]).set_title("Boxplot of tips by Day of the Week")

g = sns.FacetGrid(tips_data, row = "day")
g = g.map(plt.hist, "tip")
plt.show()
```
